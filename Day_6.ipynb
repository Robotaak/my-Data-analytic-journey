```python
import numpy as np
import pandas as pd

# Load cleaned data
df = pd.read_csv('Financials.csv')
df.columns = df.columns.str.strip()
money_cols = ['Units Sold', 'Manufacturing Price', 'Sale Price',
              'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']
for col in money_cols:
    df[col] = df[col].astype(str).str.replace(r'[$,\s-]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

print("✅ Ready. Shape:", df.shape)
```

```python
profit = df['Profit'].values  # convert to NumPy array

print("=" * 45)
print("        STATISTICAL SUMMARY")
print("=" * 45)
print(f"  Mean Profit:      ${np.mean(profit):>12,.2f}")
print(f"  Median Profit:    ${np.median(profit):>12,.2f}")
print(f"  Std Deviation:    ${np.std(profit):>12,.2f}")
print(f"  Min Profit:       ${np.min(profit):>12,.2f}")
print(f"  Max Profit:       ${np.max(profit):>12,.2f}")
print(f"  Total Profit:     ${np.sum(profit):>12,.2f}")
print(f"  25th Percentile:  ${np.percentile(profit, 25):>12,.2f}")
print(f"  75th Percentile:  ${np.percentile(profit, 75):>12,.2f}")
print("=" * 45)
```

```python
# Month over month profit growth — this is what clients LOVE
monthly = df.groupby('Month Number')['Profit'].sum().values

growth = []
for i in range(1, len(monthly)):
    pct = ((monthly[i] - monthly[i-1]) / monthly[i-1]) * 100
    growth.append(pct)

months = ['Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
print("📈 Month-over-Month Profit Growth:")
print("-" * 35)
for m, g in zip(months, growth):
    arrow = "🟢" if g > 0 else "🔴"
    print(f"  {m}: {arrow} {g:+.1f}%")

print(f"\n  Average Growth: {np.mean(growth):+.1f}%")
print(f"  Best Month Jump: {np.max(growth):+.1f}%")
print(f"  Worst Month Drop: {np.min(growth):+.1f}%")
```

```python
# How much better/worse is each segment vs company average?
avg_profit = df['Profit'].mean()
segment_avg = df.groupby('Segment')['Profit'].mean()

print(f"Company Average Profit per Deal: ${avg_profit:,.2f}")
print("\n📊 Segment Performance vs Average:")
print("-" * 45)
for seg, val in segment_avg.items():
    diff = ((val - avg_profit) / avg_profit) * 100
    arrow = "🟢" if diff > 0 else "🔴"
    print(f"  {seg:<20} {arrow} {diff:+.1f}% vs average")
```
